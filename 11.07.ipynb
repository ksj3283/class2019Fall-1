```python
#numpy: list와 달리 계산 가능.
#pure tone의 합으로 complex tone이 만들어진다. sinusoidal: sine cosine 등 sine wave. 그것들을 만드는 것을 phasor라고 함.
#사인 코사인을 쓰기 위해선 0도에서 360도를 radian으로 표현해야한다. 즉(0 - 2π. ex π/4 = 45도)
#sin0=0=sinπ cos0=1=cosπ,(sin cos 그래프 참고!)
#0-2π마다 주기 반복. 문제: 100π까지 몇 번의 주기?: 50번
#cos(θ)=θ=3/2π  cf 오일러 공식 e: 자연상수 = 2.7xxx i: 허수. 즉 e와 i는 고정된 값이고 θ(또는 t)는 변수. θ의 변화에 따라 값이 달라짐.
#f(t) = e의 ti제곱 = a+bi(임의의 복소수) = cos(t)+sin(t)i
#이때 t=0, π/2, π, 3π/2, 2π를 차례로 대입하면
#f(t)=1,i,-1,-i,1(반복) ######중요!!!왜냐하면 cos값은 1,0,-1,0 sin값은 0,1,0,-1의 순서로 반복

#복소수a+bi의 표현 방법:complex plain(복소평면) 1-> a=1 b=0, i-> a=0 b=1, -1-> a=-1 b=0, -i-> a=0 b=-1이를 xy좌표상에 옮기면
# (1,0), (0,1), (-1.0), (0,-1), 이렇게 반지름 1의 원을 이룬다.
#이때, 이 원 위의 임의의 좌표를 찍으면 부채꼴의 중심각 크기를 구할 수 있다.
#실수부분만 추출하고 싶으면 x축 값 또는 a값만 보면 됨.(위에서 projection)
#허수부만 추출하고 싶으면 y축 값 또는 b값만 보면 됨.(옆에서 projection)
```

# phasor

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.axes_grid1 import make_axes_locatable
import IPython.display as ipd
import numpy as np
%matplotlib notebook
from scipy.signal import lfilter
```

```python
# 주파수는 1초에 몇번 왔다갔다하는지 나타내는데, 단순히 phasor를 추출하면 시간의 개념이 들어있지 않다. 즉, 단순한 각도 값만 구한 것이다.
# 소리라는 것은 시간의 개념이 들어가야 실체가 된다.
```

```python
# 소리를 표현하기 위해 우리가 알기 쉽게 변수들을 설정한 것.
amp = 1                  #range
sr = 10000               #sampling rate, Hz
dur = 0.5                #in seconds
freq = 100.0             #sine frequency, Hz
```

```python
#####
#t 0.0001 0.0002 0.0003 ... 0.5000 #sampling rate가 10000이므로 1초를 10000개로 나눠야하며, duration이 0.5초이므로, 0.5000까지 가는 것.
t = np.arange(1,sr*dur+1)/sr #1-5000까지의 계산 가능한 데이터를 만들고(sampling rate와 duration을 반영하여, 1/10000초 - 5000/10000초까지 만듦)
```

```python
t
```

```python
#time이 있다고 끝난 게 아니다. 각도 값을 time과 연동시켜서 phasor로 바꿔야한다.
#generate phase
theta = t*2*np.pi * freq #t*2파이는 딱 한 바퀴 도는 것. frequency를 반영하여 100바퀴를 돌려야함. 즉 freq는 몇바퀴를 돌리느냐?를 정함.
```

```python
s=np.sin(theta)
```

```python
#문제!: time의 벡터 사이즈와 theta의 벡터 사이즈는 같다??-> 같다.
```

```python
#소리에는 시간의 개념이 필수적이다. 이제 시간의 개념을 넣어서 만들어보자.
t = np.arange(1, sr*dur+1)/sr #첫번째 time부터 1초까지의 time을 만든 것. 즉 dur
```

```python
theta = t*2*np.pi*freq
```

```python
s = np.sin(theta)  #이번에는 위와 달리 time s theta를 모두 준비한 것. 다른 것들은 고정. theta만 변수
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)             #2차원 입력. 밑의 3차원 복소수 입력과 비교!
ax.plot(t[0:1000], s[0:1000], '.')    #여기서 대괄호는 t내에서 0-1000데이터만 추출해 쓴다는 것. 여기서 t와 s의 개수는 correspond 해야함.
ax.set_xlabel('time (s)')             #시험 여기서 점들의 개수는? 답 1000개
ax.set_ylabel('real')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(t, s, '.')   
ax.set_xlabel('time (s)')             
ax.set_ylabel('real')
```

```python
#복소수의 벡터까지 추출하여 만드는 법.
c = np.exp(theta*1j)  #1j는 허수의 i. 즉 허수의 개념을 도입한 것
c
```

```python
fig = plt.figure()                           #이 모든 것의 전제가 되는 오일러 공식 명심!
ax = fig.add_subplot(111, projection='3d')   #3차원으로 바꾼 것.
ax.plot(t[0:1000], c.real[0:1000], c.imag[0:1000], '.')  #3차원의 형태. 3개의 입력, 즉 좌표 상의 한 점이(a,b,c)의 형태의 벡터로 나옴.각 벡터의 size는 correspond
ax.set_xlabel('time (s)')       #점의 개수는 위와 동일하게 1000개.
ax.set_ylabel('real')  #실수. 결과값 a+bi 중에서 a만 추출하는 역할
ax.set_zlabel('imag')  #허수  결과값 a+bi 중에서 b만 추출하는 역할. 즉 a b 값을 따로따로 받아와서 plot하는 것
```
